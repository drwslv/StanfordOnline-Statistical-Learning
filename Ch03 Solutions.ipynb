**(2)** Carefully explain the differences between the KNN classifier and KNN regression methods.

**KNN classifier:** To estimate the class of a point ing space, the class of the $K$ nearest neighboring observations is considered, and the most prevalent class is used as a prediction for the class of the point of space under consideration.

**KNN regression:** Similar to the KNN classifier, but for quantitative data. An estimate for $\hat{y}_i$ is found by considering the $K$ nearest neighbors to $x_i$ (for however many $x$s are included in the linear model), and averaging their values.

Show that we can write

What is $a_j$?

*Note: We interpret this result by saying that the fitted values from linear regression are linear combinations of the response values.*


$ \hat{y}_i = x_i \sum^{n}_{j=1} x_j y_j / \sum^{n}_{k=1} x^2_k $

$ \hat{y}_i = \sum^{n}_{j=1}  x_i x_j y_j / \sum^{n}_{k=1} x^2_k $

$ \hat{y}_i = \sum^{n}_{j=1} ( x_i x_j  / \sum^{n}_{k=1} x^2_k ) y_j$

**(7)** It is claimed in the text that in the case of simple linear regression of Y onto X, the $R^2$ statistic (3.17) is equal to the square of the correlation between X and Y (3.18). Prove that this is the case. For simplicity, you may assume that $ \bar{x} = \bar{y} = 0 $.


$$ Cor(X, Y) = \frac{ \sum^n_{i=1} (x_i - \bar{x})(y_i - \bar{y})}{ \sqrt( \sum^n_{i=1} (x_i - \bar{x})^2 ) \sqrt( \sum^n_{i=1} (y_i - \bar{y})^2 )} $$

$$ Cor(X, Y)^2 = \frac{ \sum^n_{i=1} (x_i - \bar{x})(y_i - \bar{y}) \sum^n_{i=1} (x_i - \bar{x})(y_i - \bar{y}) }{  \sum^n_{i=1} (x_i - \bar{x})^2  \sum^n_{i=1} (y_i - \bar{y})^2 } $$

$$ Cor(X, Y)^2 = \frac{ \sum^n_{i=1} (x_i - \bar{x})(y_i - \bar{y}) \sum^n_{i=1} (x_i - \bar{x})(y_i - \bar{y}) }{  \sum^n_{i=1} (x_i - \bar{x})(x_i - \bar{x})  \sum^n_{i=1} (y_i - \bar{y})(y_i - \bar{y}) } $$

Using assumption that $ \bar{x} = \bar{y} = 0 $

$$ Cor(X, Y)^2 = \frac{ \sum^n_{i=1} x_i y_i \sum^n_{i=1} x_i y_i }{  \sum^n_{i=1} x_i x_i \sum^n_{i=1} y_i y_i } = \beta \frac{ \sum^n_{i=1} x_i y_i }{ \sum^n_{i=1} y_i y_i } $$

$ \beta_0 = \bar{y} - \beta_1 \bar{x} = 0 $

$$ R^2 = \frac{TSS - RSS}{TSS} = \frac{\sum^n_{i=1} y_i y_i - \sum^n_{i=1} (y_i - \hat{y}_i)^2}{\sum^n_{i=1} y_i y_i}$$

$$ R^2 = \frac{\sum^n_{i=1} y_i y_i - \sum^n_{i=1} (y_i - \beta x_i)^2 }{\sum^n_{i=1} y_i y_i}$$

$$ R^2 = \frac{\sum^n_{i=1} y_i^2 - \sum^n_{i=1} y_i^2 - 2 \beta y_i x_i + \beta^2 x_i^2 }{\sum^n_{i=1} y_i y_i}$$

$$ R^2 = \frac{\sum^n_{i=1} y_i^2 - \sum^n_{i=1} y_i^2 + \sum^n_{i=1}  2 \beta y_i x_i - \beta^2 x_i^2 }{\sum^n_{i=1} y_i y_i}$$

$$ R^2 = \beta ( \frac{ \sum^n_{i=1}  2 y_i x_i - \beta x_i^2 }{\sum^n_{i=1} y_i y_i} )$$

$$ R^2 = \beta ( \frac{ \sum^n_{i=1}  2 y_i x_i }{\sum^n_{i=1} y_i y_i} - \frac{  \sum^n_{i=1} \beta x_i^2 }{\sum^n_{i=1} y_i y_i} )$$

$$ R^2 = \beta ( \frac{ \sum^n_{i=1}  2 y_i x_i }{\sum^n_{i=1} y_i y_i} - \beta \frac{  \sum^n_{i=1}  x_i^2 }{\sum^n_{i=1} y_i y_i} )$$

$$ R^2 = \beta ( \frac{ \sum^n_{i=1}  2 y_i x_i }{\sum^n_{i=1} y_i y_i} - \frac{ \sum^n_{i=1}  x_i y_i  \sum^n_{i=1}  x_i^2 }{ \sum^n_{i=1}  x_i^2 \sum^n_{i=1} y_i y_i} )$$

$$ R^2 = \beta ( \frac{ \sum^n_{i=1}  2 y_i x_i }{\sum^n_{i=1} y_i y_i} - \frac{ \sum^n_{i=1}  x_i y_i }{ \sum^n_{i=1} y_i y_i} )$$

$$ R^2 = \beta \frac{ \sum^n_{i=1}  x_i y_i }{\sum^n_{i=1} y_i y_i} = Cor(X, Y)^2 $$

```python
def abline(ax, b, m, *args, **kwargs):
    "Add a line with slope m and intercept b to ax"
    xlim = ax.get_xlim()
    ylim = [m * xlim[0] + b, m * xlim[1] + b] # predicted values for any particular set of x
    ax.plot(xlim, ylim, *args, **kwargs)
```

This problem involves simple linear regression without an intercept.

**(a)** Recall that the coefficient estimate $ \hat{\beta} $ for the linear regression of Y onto X without an intercept is given by (3.38). Under what circumstance is the coefficient estimate for the regression of X onto Y the same as the coefficient estimate for the regression of Y onto X?

The coefficient is the same when $ \hat{\beta} = 1/ \hat{\beta} $, so when $ \hat{\beta} = 1, -1 $.

This is true when $ Cor(X,Y) = 1, -1 $, and because

$$ Cor(X,Y) = \frac{Cov(X,Y)}{\sigma_X \sigma_Y} $$

this will be true when

$$ Cov(X,Y) = \sigma_X \sigma_Y $$

Or from the equation

$$ Cor(X,Y) = \frac{ \sum^n_{i=1} (x_i - \bar{x}) (y_i - \bar{y})  }{ \sqrt{ \sum^n_{i=1}  ( x_i - \bar{x}  )^2 }  \sqrt{ \sum^n_{i=1}  ( y_i - \bar{y}  )^2  }  } $$

implies

$$ \sum^n_{i=1} (x_i - \bar{x}) (y_i - \bar{y})  =  \sqrt{ \sum^n_{i=1}  ( x_i - \bar{x}  )^2 }  \sqrt{ \sum^n_{i=1}  ( y_i - \bar{y}  )^2  }       $$

$$ \beta = \frac{ \sum^n_{i=1} (x_i - \bar{x}) (y_i - \bar{y}) }{ \sum^n_{i=1} (x_i - \bar{x})^2  } = \frac{ \sqrt{ \sum^n_{i=1}  ( x_i - \bar{x}  )^2 }  \sqrt{ \sum^n_{i=1}  ( y_i - \bar{y}  )^2  }  }{ \sum^n_{i=1} (x_i - \bar{x})^2  }     $$

$$ \beta = \frac{  \sqrt{ \sum^n_{i=1}  ( y_i - \bar{y}  )^2  }  }{  \sqrt{ \sum^n_{i=1}  ( x_i - \bar{x}  )^2 }  }     $$

If $ \bar{y} = \bar{x} = 0 $ (which should be true of regressing without an intercept)$

$$ \beta = \frac{  \sqrt{ \sum^n_{i=1}  y_i^2  }  }{  \sqrt{ \sum^n_{i=1}  x_i^2 }  }     $$

$ \beta $ must be = 1, so

$$ \sum^n_{i=1} x_i^2 =  \sum^n_{i=1} y_i^2 $$

Better version:

$$ \beta_x = \frac{ \sum^n_{i=1} x_i y_i }{ \sum^n_{i=1} x_i^2  } = \beta_y = \frac{ \sum^n_{i=1} x_i y_i }{ \sum^n_{i=1} y_i^2  } $$

This is true when

$$ \sum^n_{i=1} x_i^2 =  \sum^n_{i=1} y_i^2 $$

SO, THE TRUE SOLUTON $ \beta = 1 $ FOR TRUE BETA, but $ \hat{\beta} $ can actually be symmetrical for many estimates. It just won't be significant if not close to 1, it seems (becuase it's a bad estimate of the true value).
